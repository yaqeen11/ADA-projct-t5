Importing packages for analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sna
```

Importing the data file

```python
df=pd.read_csv('sels.csv')
```

Top 5 rows of the dataset

```python
df.head()
```

```python
# التعرف على أنواع البيانات وكم تحتوي على rows and columns
df.info()
```

```python
#returns the number of missing values in each column. معرفة البيانات المفقودة في كل عامود
df.isna().sum()
```

```python
#will drop all missing values في السنة والناشر 
df.dropna(inplace=True)
df.isna() .sum()
```

```python
#Decriptive Statistics of the dataset
df.describe()
```

```python
#Finding out the game with maximum sales globally
df[df['Global_Sales'] == df['Global_Sales'].max()]
```

```python
#Finding out the game with minimum sales globally
df[df['Global_Sales'] == df['Global_Sales'].min()]
```

```python
#I have a dataframe of taxi data with two columns بين السنوات والمبيعات العالمية
df.groupby('Year').Global_Sales.value_counts()
```

```python
#Global Sales over the years
plt.figure(figsize=[9,5])
type_of_travel=df.Year
very_frq=df['Global_Sales']
plt.grid()
plt.bar(type_of_travel,very_frq,color='g');
plt.xlabel('Types of travel',fontsize = 15)
plt.ylabel('Preference number',fontsize = 15);
plt.title('Show of the preferred type ',fontsize = 20, weight = 'bold');
```

```python
#I have a dataframe of taxi data with two columns بين السنوات والمبيعات في اليابان
df.groupby('Year').JP_Sales.value_counts()
```

```python
#Sales of different genre in North America
ax=plt.figure(figsize=(15,6))
sna.barplot(x='Genre',y='NA_Sales',data=df,palette='Paired')
plt.ylabel('Sales')
plt.title('Sales of different genre in North America')
```

```python
#What is the most popular game name المشهورة وكم مره تكرار لها 
df.Name.value_counts()
```

```python
#Sales of games in different platforms globally
ax=plt.figure(figsize=(15,6))
df.groupby(['Platform'])['Global_Sales'].sum().plot.bar()
plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.title('Sales of games in different platforms globally')
```
